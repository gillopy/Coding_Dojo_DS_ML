### Importamos las librerias necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

### Cargamos el Dataset y hacemos las primeras impresiones

```python
df = pd.read_csv('World-happiness-report-2024.csv')
```

```python

print(df.shape)
print('-----------------------------')
print(df.info())
print('-----------------------------')
df.head()
```

### Verificamos si hay nulos

```python
df.isnull().sum()
```

Eliminamos las 3 filas con nulos porque no afecta el analisis

```python
# Eliminar filas con datos faltantes
df = df.dropna()
```

Verificamos cuales son nuestras columnas

```python
df.columns
```

Creamos los primeros histogramas

```python
plt.figure()
sns.histplot(df['Ladder score'], color='skyblue',kde=True, edgecolor='black')
plt.grid()
plt.show()
```

```python
plt.figure()
sns.histplot(df['Log GDP per capita'], color='skyblue',kde=True, edgecolor='black')
plt.grid()
plt.show()
```

```python
plt.figure()
sns.histplot(df['Healthy life expectancy'], color='skyblue',kde=True, edgecolor='black')
plt.grid()
plt.show()
```

```python
plt.figure()
sns.histplot(df['Freedom to make life choices'], color='skyblue', kde=True, edgecolor='black')
plt.grid()
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Ladder score'])
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Log GDP per capita'])
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Healthy life expectancy'])
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Freedom to make life choices'])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True, gridspec_kw={'height_ratios': [1, 4]})

# Dibujar el boxplot en el primer subplot (arriba)
sns.boxplot(x=df['Ladder score'], ax=ax1, color='lightgray', width=0.5)
ax1.set(xlabel='')

# Dibujar el histograma en el segundo subplot (abajo)
sns.histplot(df['Ladder score'], color='skyblue', kde=True, edgecolor='black', ax=ax2)

# Ajustar los espacios entre los gráficos
plt.tight_layout()
plt.grid()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True, gridspec_kw={'height_ratios': [1, 4]})

# Dibujar el boxplot en el primer subplot (arriba)
sns.boxplot(x=df['Log GDP per capita'], ax=ax1, color='lightgray', width=0.5)
ax1.set(xlabel='')

# Dibujar el histograma en el segundo subplot (abajo)
sns.histplot(df['Log GDP per capita'], color='skyblue', kde=True, edgecolor='black', ax=ax2)

# Ajustar los espacios entre los gráficos
plt.tight_layout()
plt.grid()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True, gridspec_kw={'height_ratios': [1, 4]})

# Dibujar el boxplot en el primer subplot (arriba)
sns.boxplot(x=df['Healthy life expectancy'], ax=ax1, color='lightgray', width=0.5)
ax1.set(xlabel='')

# Dibujar el histograma en el segundo subplot (abajo)
sns.histplot(df['Healthy life expectancy'], color='skyblue', kde=True, edgecolor='black', ax=ax2)

# Ajustar los espacios entre los gráficos
plt.tight_layout()
plt.grid()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8), sharex=True, gridspec_kw={'height_ratios': [1, 4]})

sns.boxplot(x=df['Freedom to make life choices'], ax=ax1, color='lightgray', width=0.5)
ax1.set(xlabel='')

sns.histplot(df['Freedom to make life choices'], color='skyblue', kde=True, edgecolor='black', ax=ax2)

plt.tight_layout()
plt.grid()
plt.show()

fig.savefig('ejemplo.png')
```

La distribución sesgada a la derecha muestra que la libertad para tomar decisiones de vida no está distribuida de manera uniforme entre los países. La mayoría de los países reportan niveles moderados de libertad, pero un pequeño grupo de países tienen muy poca libertad, lo que se destaca con los valores atípicos visibles en el boxplot.
