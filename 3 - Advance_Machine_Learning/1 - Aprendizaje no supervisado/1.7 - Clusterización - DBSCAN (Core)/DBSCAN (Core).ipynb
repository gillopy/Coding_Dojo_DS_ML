### **Preprocesamiento**

Creamos una copia el dataset para trabajar mejor

```python
data2 = data.copy()
```

```python
data2.head()
```

#### **Imputacion de Outliers por mediana**

```python
import pandas as pd

# Supongamos que 'data2' ya está cargado
# Iterar sobre todas las columnas numéricas
for columna in data2.select_dtypes(include='number').columns:
    Q1 = data2[columna].quantile(0.25)  # Primer cuartil
    Q3 = data2[columna].quantile(0.75)  # Tercer cuartil
    IQR = Q3 - Q1  # Rango intercuartil

    # Límites para definir outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Reemplazar valores fuera de los límites por la mediana
    mediana = data2[columna].median()
    data2[columna] = data2[columna].apply(
        lambda x: mediana if x < limite_inferior or x > limite_superior else x
    )
```

#### **Escalamos**

```python
# Escalamos los datos
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data2)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled
```

#### **DBSCAN**

```python
epsilon = 0.3  # Ajusta según las nuevas dimensiones
min_samples = 3

dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
clusters = dbscan.fit_predict(data_scaled)

# Agregar resultados al dataset original
data['Cluster'] = clusters

# Generar paleta dinámica
num_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)  # Excluir outliers (-1)
palette = sns.color_palette("Set2", num_clusters)  # Crear paleta con suficientes colores

# Visualización
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data['Milk'], y=data['Grocery'],
    hue=data['Cluster'],
    palette=palette + ['red'],  # Agregar rojo explícito para los outliers
    style=(data['Cluster'] == -1).map({True: 'X', False: 'o'}),  # Diferenciar con marcadores
    s=100
)
plt.title("Clusters formados por DBSCAN (outliers en rojo)")
plt.xlabel("Milk")
plt.ylabel("Grocery")
plt.legend(title="Cluster", loc="upper right")
plt.show()

# Estadísticas de los clústeres
print(data['Cluster'].value_counts())
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    data['Milk'], data['Grocery'], data['Fresh'],
    c=data['Cluster'], cmap='Set2', s=100, alpha=0.8
)

ax.set_title("Clusters formados por DBSCAN (3D)")
ax.set_xlabel("Milk")
ax.set_ylabel("Grocery")
ax.set_zlabel("Fresh")
plt.show()
```
